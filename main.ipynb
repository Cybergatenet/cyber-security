### Write a program to sort a list without using the SORTED method

```python
def mySort(arr):
    # Nested Loop
    answer = []
    for i in arr:
        for j in arr:
            # print(i, j)
            if i > j:
                i = j
                answer.append(i)
                # answer += j
    print(answer)


def mySort(arr):
    n = len(arr)

    for i in range(n):
        min_index = i

        for j in range(i + 1, n):
            if arr[j] < arr[min_index]:
                min_index = j

        # Swap
        arr[i], arr[min_index] = arr[min_index], arr[i]

    return arr

# abel_age = 34
# presh_age = 23

# abel_age, presh_age = 34, 23
```
